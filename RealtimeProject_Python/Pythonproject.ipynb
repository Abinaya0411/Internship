```python
import pandas as pd
```

```python
dataset=pd.read_csv("PreInputdata.csv")
```

```python
dataset
```

```python
dataset.rename(columns={'Day': 'Date', 'Campaign name': 'Campaign name', 'Amount spent (INR)': 'Ad Spent', 
                  'Results': 'No. of leads', 'CTR (link click-through rate)': 'CTR', 'Link clicks': 'Link click', 
                  'Landing page views': 'LP views'},inplace=True)
```

```python
dataset
```

```python
output_data = dataset[['Date','Campaign name','Ad Spent','No. of leads','CTR','Link click','LP views']]
```

```python
import numpy as np
output_data["Cost Per Lead"]=dataset["Ad Spent"]/dataset["No. of leads"]
output_data["Cost Per Lead"]= output_data["Cost Per Lead"].replace([np.inf, -np.inf], np.nan)
output_data["Cost Per Lead"].fillna(0, inplace= True)
```

```python
output_data["Member passed"]=(output_data["LP views"]/output_data["Link click"])*100
output_data["Member passed"].fillna(0, inplace= True)
```

```python
output_data["LP Conversion %"]=(output_data["No. of leads"]/output_data["LP views"])*100
output_data["LP Conversion %"]= output_data["LP Conversion %"].replace([np.inf, -np.inf], np.nan)
output_data["LP Conversion %"].fillna(0, inplace= True)
```

```python
output_data['Date'] = pd.to_datetime(output_data['Date']).dt.strftime('%d %b (%a)')
```

```python
output_data.sort_values(by=['Campaign name','Date'])
```

```python
output_data=output_data.reset_index(drop=True)
```

```python
output_data
```

```python
output_data.to_csv("Output.csv",index=False)
```
