```python
import pandas as pd
```

```python
dataset=pd.read_csv("Inputdata - Sheet1.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtypes=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual= quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='constant', fill_value=0)
imp.fit(dataset[quan])
quan_data=imp.transform(dataset[quan])
```

```python
quan_dataset=pd.DataFrame(quan_data,columns=quan)
```

```python
quan_dataset
```

```python
quan_dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='constant', fill_value='missing_value')
imp.fit(dataset[qual])
qual_data=imp.transform(dataset[qual])
```

```python
qual_dataset=pd.DataFrame(qual_data,columns=qual)
```

```python
qual_dataset
```

```python
qual_dataset.isnull().sum()
```

```python
final_dataset=pd.concat([qual_dataset,quan_dataset],axis=1)
```

```python
final_dataset.reset_index(drop=True, inplace=True)
```

```python
final_dataset
```

```python
final_dataset.isnull().sum()
```

```python
final_dataset.to_csv("PreInputdata.csv",index=False)
```
